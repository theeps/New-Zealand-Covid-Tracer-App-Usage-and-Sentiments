## MERGE IOS and GOOGLE PLAY STORE REVIEWS

```python
frames=[dfggle,dfios]
df1=pd.concat(frames)
df1.head(5)
```

```python
# applying ffill() method to fill the missing values
col = ['Version']
df1.loc[:,col] = df1.loc[:,col].bfill()
df1.head()
```

```python
df1.dropna(inplace=True)
df1.shape
```

```python
list=['Rating']
for i in list:
    print('{} has unique values as {}'.format(i,df1[i].unique()))
```

```python
#ratings= ['Negative','Positive','Neutral']
ratings= [1,2,3,4,5]
df1 = df1.loc[df1['Rating'].isin(ratings)]
df1.head(40)
```
