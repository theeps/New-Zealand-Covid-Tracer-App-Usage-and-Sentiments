```python
import matplotlib.dates as mdates
plt.style.use('default')
df1['Date'] = pd.to_datetime(df1['Date/Time To(New)'])
df1['App Registrations'] = pd.to_numeric(df1['App Registrations'])
# date1 = 

# Plot
ax = plt.figure(figsize=(16,9))
ax = plt.plot(df1['Date'], df1['App Registrations'])
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
ax = plt.gcf().autofmt_xdate()# Rotation

# plt.axvline(dt.datetime(2020,8, 2),ymin=0, ymax=210000, color='r', label='test lines')
# ax.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.axvline(pd.Timestamp('2020-08-11'),color='#8B1A1A',linestyle='--', label='11-08-2020')
plt.axvline(pd.Timestamp('2021-01-24'),color='r',linestyle='--', label='24-01-2021')
plt.axvline(pd.Timestamp('2021-02-14'),color='#FF8000',linestyle='--', label='14-02-2021')
plt.axvline(pd.Timestamp('2021-02-28'),color='#3B3B3B',linestyle='--', label='28-02-2021')
plt.axvline(pd.Timestamp('2021-08-17'),color='g',linestyle='--', label='17-08-2021')
plt.axvline(pd.Timestamp('2021-09-07'),color='indigo',linestyle='--', label='07-09-2021')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left',title = "Significant Dates")
plt.grid(b=True,color='gray')
#plt.title("App Registrations through Time",loc='left')
plt.xlabel("Date")
plt.ylabel("App reg per day")
```
