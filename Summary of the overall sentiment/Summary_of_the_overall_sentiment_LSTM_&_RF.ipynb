#Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
%matplotlib inline
```

## **LSTM**

```python
LSTM=pd.read_csv('/content/nzpredicted_DL2.csv')
LSTM.tail(5)
```

Count of sentiments

```python
# Let's count the number of sentiments
sentiment_counts = LSTM.groupby(['predicted_label_name']).size()
print(sentiment_counts)

# Let's visualize the sentiments
fig = plt.figure(figsize=(4,4), dpi=100)
ax = plt.subplot(111)
sentiment_counts.plot.pie(ax=ax, autopct='%1.1f%%', startangle=270, fontsize=12, label="")
```

Count of Ratings

```python
# Let's count the number of sentiments for ratings converted
sentiment_counts = LSTM.groupby(['Rating_converted']).size()
print(sentiment_counts)

# Let's visualize the sentiments
fig = plt.figure(figsize=(4,4), dpi=100)
ax = plt.subplot(111)
sentiment_counts.plot.pie(ax=ax, autopct='%1.1f%%', startangle=270, fontsize=12, label="")
```

Counts of sentiments which are correctly classified according to the ratings

```python
# Let's count the number of sentiments correctly classified
sentiment_counts = LSTM.groupby(['check']).size()
print(sentiment_counts)

# Let's visualize the sentiments
fig = plt.figure(figsize=(4,4), dpi=100)
ax = plt.subplot(111)
sentiment_counts.plot.pie(ax=ax, autopct='%1.1f%%', startangle=270, fontsize=12, label="")
```
